```python
for i in range(1,1000):
    try:
        url="https://www.prokabaddi.com/sifeeds/kabaddi/static/json/"+str(i)+"_player.json"
        response=json.loads(requests.get(url).text)
        df_player_tmp=json_normalize(response['over_all_stats'])
        #print(df_player_tmp)
        #print(df_player_tmp.type())
        frames = [df_player,df_player_tmp]
        df_player=pd.concat(frames)
    except Exception as e:
        print("No player:"+str(i))
df_player.head(100)
```
